# 2. Encoding Techniques:

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
data1=data.select_dtypes(include=['object']).columns
```

```python
data1
```

```python
encoded_df = data.copy()
for column in data1:
    unique_values = encoded_df[column].nunique()
    if unique_values <= 5:
        # Apply One-Hot Encoding
        encoded_df = pd.get_dummies(encoded_df, columns=[column], drop_first=True)
    else:
        # Apply Label Encoding
        le = LabelEncoder()
        encoded_df[column] = le.fit_transform(encoded_df[column])
encoded_df.head()
```

# •	Discuss the pros and cons of One-Hot Encoding and Label Encoding.

```python
   ***One-Hot Encoding
Pros:
   No Ordinal Relationship: Suitable for nominal data where there is no natural order between categories.
Model Compatibility: Works well with algorithms that require input data without ordinal relationships (e.g., linear regression, KNN).
Cons:
   High Dimensionality: Increases the number of features significantly, especially with variables that have many categories.
Memory and Computational Cost: More memory and computation are required due to the increased number of columns.

   ***Label Encoding
Pros:
Simplicity and Efficiency: Converts categories to numerical values directly, which is memory efficient.
No Increase in Dimensionality: Keeps the number of features unchanged.
Cons:
   Implied Ordinality: Imposes an ordinal relationship between categories, which may not be appropriate for nominal data.
```

# Explain the rationale behind your choices.

```python
The rationale behind choosing specific scaling techniques depends on various factors related to the datasets characteristics 
and the requirements of the machine learning algorithm being used.

*Rationale:*
Range Sensitivity: Machine learning algorithms based on distance calculations perform better when the features are within a similar range.
If features are not scaled, those with larger ranges can dominate the learning process, leading to suboptimal performance.
Preservation of Relationships: Min-Max scaling maintains the original distribution of the data,
only compressing or expanding the values.
```
