# Task 4: SVM Implementation

```python
from sklearn.svm import SVC
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
svm_classifier = SVC(kernel='linear', random_state=42)
```

```python
svm_classifier.fit(X_train,y_train)
```

```python
y_pred = svm_classifier.predict(X_test)
y_pred 
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1-Score: {f1:.4f}')
```

```python
X_train_2d = X_train.iloc[:, :2]
X_test_2d = X_test.iloc[:, :2]
```

```python
svm_classifier_2d = SVC(kernel='linear', random_state=42)
svm_classifier_2d.fit(X_train_2d, y_train)
```

```python
def plot_decision_boundary(model, X, y):
    x_min, x_max = X.iloc[:, 0].min() - 1, X.iloc[:, 0].max() + 1
    y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
    plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, s=20, edgecolor='k', cmap='coolwarm')
    plt.title('Decision Boundary')
    plt.xlabel(X.columns[0])
    plt.ylabel(X.columns[1])
    plt.show()

```

```python
plot_decision_boundary(svm_classifier_2d, X_train_2d, y_train)
```

```python
plot_decision_boundary(svm_classifier_2d, X_test_2d, y_test)
```

```python
kernels = ['linear', 'poly', 'rbf']
results = {}
for kernel in kernels:
    svm_classifier = SVC(kernel=kernel, random_state=42)
    svm_classifier.fit(X_train, y_train)
    y_pred = svm_classifier.predict(X_test)
```

# 2.	Analyze SVM strengths and weaknesses for the Mushroom dataset based on EDA and visualization results.

```python
*Strengths of SVM for the Mushroom Dataset*
1.Good at Handling Non-Linear Data: SVM can use different kernels to handle complex, non-linear relationships in the data. This is helpful if the mushroom dataset features aren’t linearly separable.
   
2.Works Well with Many Features: If the dataset has lots of features (like different attributes of mushrooms), SVM can still perform effectively without getting overwhelmed.

3.Clear Separation: SVM focuses on finding the best boundary that separates the classes (edible vs. poisonous) with the maximum margin, which usually leads to better classification.

*Weaknesses of SVM for the Mushroom Dataset*
1.Computationally Expensive: SVM can be slow to train, especially if the dataset is large, which might be an issue with a big mushroom dataset.

2.Needs Careful Tuning: The performance of SVM depends heavily on choosing the right settings (like the kernel type and regularization parameters). This can be tricky and time-consuming.

3.Hard to Interpret: SVM models, especially with non-linear kernels, are not easy to interpret. Understanding why the model made a certain prediction might be difficult.

```
