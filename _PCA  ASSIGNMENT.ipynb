# 3. Clustering with Original Data:

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
clusters=KMeans(n_clusters=4,random_state=42)
kmeans=clusters.fit_predict(data_normal)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data_normal[:, 0], data_normal[:, 1], c=kmeans, cmap='viridis', marker='o')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.title('K-means Clustering of Wine Dataset (First Two Features)')
plt.show()
```

```python
silhouette_avg = silhouette_score(data_normal, kmeans)
print(f'Silhouette Score for K-means: {silhouette_avg}')
```

# 4. Clustering with PCA Data:

```python
cluster_pca=KMeans(n_clusters=4,random_state=42)
kmeans_pca=cluster_pca.fit_predict(pca_values)
kmeans_pca
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(pca_values[:, 0], pca_values[:, 1], c=kmeans_pca, cmap='viridis', marker='o')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.title('K-means Clustering of pca(First Two Features)')
plt.show()
```

```python
silhouette_pca= silhouette_score(pca_values, kmeans_pca)
print(f'Silhouette Score for K-means: {silhouette_pca}')
```

# 5.Comparison and Analysis:

```python
1) The original dataset scatter plot uses the first two features, which may not capture the most variance.The PCA-transformed scatter plot uses the first two principal components, which capture the most varianc

2) Both methods aim to cluster data points into meaningful groups. Both visualizations can show clear clusters if the data structure is strong.

3) PCA reduces dimensionality, retaining the most important features.It can improve clustering performance by removing noise and redundantfeatures.
```

# 6. Conclusion and Insights

```python
1)Applying PCA before clustering can significantly improve clustering performance by reducing noise and redundancy in the dataset.

2) PCA simplifies the dataset by reducing the number of features while retaining the most important information. This is especially useful for high-dimensional datasets where many features may be redundant or noisy.

3)Use PCA when dealing with high-dimensional data or when clustering performance needsenhancement.
```
