# ASSOCIATION RULES

```python
data=pd.read_excel("C:\\Users\\Admin\\AppData\\Local\\Temp\\Rar$DIa0.393\\Online retail.xlsx")
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

transactions = data['shrimp,almonds,avocado,vegetables mix,green grapes,whole weat flour,yams,cottage cheese,energy drink,tomato juice,low fat yogurt,green tea,honey,salad,mineral water,salmon,antioxydant juice,frozen smoothie,spinach,olive oil'].apply(lambda x: x.split(','))

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
# Apply Apriori
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
# Generate the association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.2)
rules
```

```python
rules_sorted=rules.sort_values(by='lift',ascending=False)
rules_sorted
```

# 1.	What is lift and why is it important in Association rules?

```python
Lift is a measure used in association rule mining to evaluate the strength of an association between two items or itemsets. In simple terms, it helps to identify how much more likely it is for two items to be found together than would be expected by chance.

IMPORTANCE:
   Lift is important in association rules because it shows how much more likely two items are to occur together than by random
chance. A higher lift value indicates a stronger and more significant association between the items, helping to identify
meaningful patternsÂ inÂ theÂ data.
```

# 2.	What is support and Confidence. How do you calculate them?

```python
SUPPORT:
    Support is a number of transactions in which antecedent and consequent appear in the data
CALCULATION:
    Transaction with antecedent and consequent / Total Transactions
CONFIDENCE:
    Confidence is a number of antecedent transactions that also have the consequent item set
CALCULATION:
    Transactions with antecedent and consequent /Transaction with antecedent
```
